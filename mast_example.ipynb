```python
from astroquery.mast import Mast, Observations
from astropy.io import fits
from astropy import table as tb
from astropy.time import Time

from glob import glob
import matplotlib.pyplot as plt
import matplotlib.image as img
from tqdm.notebook import tqdm
import numpy as np

directory = '/Volumes/hdd_1/jwst_data'
dirstar = directory + '/mastDownload/JWST/**/'
```

```python
obs = Observations.query_criteria(
    obs_collection = 'JWST',
    instrument_name = 'NIRCAM/CORON',
    target_name = 'TWA-43',
    proposal_id = '4050',
    calib_level = '3'
)
print(f'Found {len(obs)} observations:')
for title in set(obs['obs_title']):
    print(f' - {title}')
```

```python
obs
```

```python
t = [Observations.get_product_list(o) for o in obs]
files = tb.unique(tb.vstack(t), keys = 'productFilename')
len(files)
```

```python
manifest = Observations.download_products(
    files,
    curl_flag=False,
    calib_level = [3],
    download_dir = directory
)
```

```python
fnames = glob(dirstar + '*.jpg')
for fname in tqdm(fnames):
    namearray = fname.split('/')[-1].split('_')
    plt.imshow(img.imread(fname))
    plt.title('_'.join(namearray[-2:]))
    plt.show()
```
